```python
import pandas as pd
```

```python
df_ad = pd.read_csv('google ads.csv')
```

```python
df_ad
```

```python
df_category = pd.read_csv('domain_with_category_fin.csv')
```

```python
df_category
```

```python
df_ad.rename(columns={'Unnamed: 1': 'On_Off'}, inplace=True)
df_ad
```

```python
# 统计用户的广告类别
# on
user_ads_on = {}
df_on = df_ad[df_ad['On_Off'] == 'ON']
df_on
```

```python
# off
user_ads_off = {}
df_off = df_ad[df_ad['On_Off'] == 'OFF']
df_off
```

```python
workers = list(set(df_on.Worker.to_list()))
for worker in workers:
    user_ads_on[worker] = []
    user_ads_off[worker] = []
```

```python
ads_list = df_category.domain.to_list()
```

```python
df_on.reset_index(drop=True, inplace=True)
df_off.reset_index(drop=True, inplace=True)
```

```python
from tqdm import tqdm

for i in tqdm(range(len(df_on))):
    for j in range(4, df_on.shape[1]):
        if (df_on.iloc[i, j] == df_on.iloc[i, j]):
            user_ads_on[df_on.loc[i, 'Worker']].append(df_on.iloc[i, j])
```

```python
for i in tqdm(range(len(df_off))):
    for j in range(4, df_off.shape[1]):
        if (df_off.iloc[i, j] == df_off.iloc[i, j]):
            user_ads_off[df_off.loc[i, 'Worker']].append(df_off.iloc[i, j])
```

```python
# on
# 看 ads 在在多少人的网页上出现过
from collections import Counter
on_ads = []
for ad in ads_list:
    for k,v in user_ads_on.items():
        if ad in v:
            on_ads.append(ad)
            continue
counter = Counter(on_ads)
top5 = counter.most_common(5)
for i in range(len(top5)):
    print("{0}: {1}".format(top5[i][0], top5[i][1]))
```

```python
# off
# 看 ads 在在多少人的网页上出现过
from collections import Counter
off_ads = []
for ad in ads_list:
    for k,v in user_ads_off.items():
        if ad in v:
            off_ads.append(ad)
            continue
counter = Counter(off_ads)
top5 = counter.most_common(5)
for i in range(len(top5)):
    print("{0}: {1}".format(top5[i][0], top5[i][1]))
```

```python
user_cate_on = {}
user_cate_off = {}
for worker in workers:
    user_cate_on[worker] = []
    user_cate_off[worker] = []
```

```python
for worker in workers:
    for ad in user_ads_on[worker]:
        try:
            ad_category = df_category[df_category['domain'] == ad].iloc[0, 2]
            user_cate_on[worker].append(ad_category)
        except:
            pass
```

```python
user_ads_on
```

```python
for worker in workers:
    for ad in user_ads_off[worker]:
        ad_category = df_category[df_category['domain'] == ad].iloc[0, 2]
        user_cate_off[worker].append(ad_category)
```

```python
user_ads_off
```

```python
# Top 5 popular categories
from collections import Counter

for k, v in user_cate_on.items():
    counter = Counter(v)
    top5 = counter.most_common(5)
    print(k)
    top5_sum = 0
    for i in range(len(top5)):
        print("{0}: {1}".format(top5[i][0], top5[i][1]))
        top5_sum += top5[i][1]
    print("({:.2%})".format(top5_sum / len(v)))
    print('---')
```

```python
# Top 5 popular categories
from collections import Counter

for k, v in user_cate_off.items():
    if v:
        counter = Counter(v)
        top5 = counter.most_common(5)
        print(k)
        top5_sum = 0
        for i in range(len(top5)):
            print("{0}: {1}".format(top5[i][0], top5[i][1]))
            top5_sum += top5[i][1]
        print("({:.2%})".format(top5_sum / len(v)))
    print('---')
```
