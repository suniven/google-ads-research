```python
plt.figure(figsize=(8, 5), dpi=120)

ax = sns.scatterplot(x='gmail account years', y='count of tags', data=df, markers=True, color="#87CEFA")

# for i, point in df.iterrows():
#     ax.text(point['gmail account years']+0.2, point['count of tags']-5, f"({point['gmail account years']:.1f}, {point['count of tags']:.0f})")

# 
for i, point in df.iterrows():
    # 
    is_overlapping = False
    for j, other_point in df.iterrows():
        if j >= i:
            break
        if abs(point['gmail account years'] - other_point['gmail account years']) < 1 and abs(
                point['count of tags'] - other_point['count of tags']) < 10:
            is_overlapping = True
            ax.text(point['gmail account years'] - 0.5, point['count of tags'] + 6,
                    f"({point['gmail account years']:.1f}, {point['count of tags']:.0f})")
            break
    # 
    if not is_overlapping:
        ax.text(point['gmail account years'] - 0.5, point['count of tags'] - 10,
                f"({point['gmail account years']:.1f}, {point['count of tags']:.0f})")

ax.set_ylabel('Count of tags', fontsize=8)
ax.set_xlabel('Gmail account age', fontsize=8)
plt.savefig('./figs/gamil_account_years-tag_count.pdf', bbox_inches='tight')
plt.show()
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        # 
        return '{p:.2f}% ({v:d})'.format(p=pct, v=val)

    return my_autopct
```

```python
gender_dic = df.gender.value_counts().to_dict()
```

```python
gender = [key for key in gender_dic]
count = [value for value in gender_dic.values()]
plt.figure(figsize=(5, 5), dpi=150)
plt.pie(count, labels=gender, autopct=make_autopct(count),
        textprops={'fontsize': 9}, labeldistance=1.15, explode=[0.04, 0])
plt.legend(fontsize=6)

plt.show()
```

```python
age_dict = {"20s": 0,
            "30s": 0,
            "40s": 0,
            "60s": 0}
for i in range(len(df)):
    age = df.loc[i, 'age']
    if age >= 20 and age < 30:
        age_dict["20s"] += 1
    elif age >= 30 and age < 40:
        age_dict["30s"] += 1
    elif age >= 40 and age < 50:
        age_dict["40s"] += 1
    # elif age >=50 and age < 60:
    #     age_dict["50s"] += 1
    elif age >= 60 and age < 70:
        age_dict["60s"] += 1
```

```python
age_dict
```

```python
# patchestexts1labeltexts2
labels = [key for key in age_dict]
count = [value for value in age_dict.values()]
plt.figure(figsize=(5, 5), dpi=120)
colors = sns.color_palette('pastel')
patches, l_text, p_text = plt.pie(count, labels=labels, autopct=make_autopct(count),
                                  textprops={'fontsize': 9}, labeldistance=1.1,
                                  pctdistance=0.6, colors=colors)
plt.legend(fontsize=8)

plt.savefig('./figs/age.pdf', bbox_inches='tight')
plt.show()
```
