```python
import pandas as pd
```

```python
domain_list = []

from tqdm import tqdm

for i in tqdm(range(len(df))):
    for j in range(4, df.shape[1]):
        if (df.iloc[i, j] == df.iloc[i, j]):
            domain_list.append(df.iloc[i, j])
```

```python
# 统计用户的广告数量
# on
user_ads_on = {}
df_on = df[df['On_Off']=='ON']
df_on
```

```python
workers = list(set(df_on.Worker.to_list()))
```

```python
for worker in workers:
    user_ads_on[worker] = []
```

```python
user_ads_on
```

```python
df_on.reset_index(drop = True,inplace = True)
```

```python
df_on
```

```python
for i in tqdm(range(len(df_on))):
    for j in range(4, df_on.shape[1]):
        if (df_on.iloc[i, j] == df_on.iloc[i, j]):
            user_ads_on[df_on.loc[i, 'Worker']].append(df_on.iloc[i, j])
```

```python
user_ads_on
```

```python
# ads count
for k,v in user_ads_on.items():
    print("{0}: {1}".format(k, len(v)))
```

```python
# unique ads
for k,v in user_ads_on.items():
    print("{0}: {1}".format(k, len(set(v))))
```

```python
# Top 5 popular ads
from collections import Counter

for k,v in user_ads_on.items():
    counter = Counter(v)
    top5 = counter.most_common(5)
    print(k)
    top5_sum = 0
    for i in range(len(top5)):
        print("{0}: {1}".format(top5[i][0], top5[i][1]))
        top5_sum += top5[i][1]
    print("({:.2%})".format(top5_sum/len(v)))
    print('---')
```

```python
# exclusive ads

for worker in tqdm(workers):
    exclusive_ads_count = 0
    print(worker)
    for ad in user_ads_on[worker]:
        exclusive = True
        for worker2 in workers:
            if not exclusive:
                break
            if worker2 == worker:
                continue
            ads2 = list(set(user_ads_on[worker2]))
            for ad2 in ads2:
                if ad == ad2:
                    exclusive = False
                    break
        if exclusive:
            print(ad)
            exclusive_ads_count += 1
    print("{0}: {1}".format(worker, exclusive_ads_count))
    print("({:.2%})".format(exclusive_ads_count/len(user_ads_on[worker])))
    print('---')
```

```python
# 统计用户的广告数量
# off
user_ads_off = {}
df_off = df[df['On_Off']=='OFF']
df_off
```

```python
for worker in workers:
    user_ads_off[worker] = []
```

```python
df_off.reset_index(drop = True,inplace = True)
df_off
```

```python
for i in tqdm(range(len(df_off))):
    for j in range(4, df_off.shape[1]):
        if (df_off.iloc[i, j] == df_off.iloc[i, j]):
            user_ads_off[df_off.loc[i, 'Worker']].append(df_off.iloc[i, j])
```

```python
user_ads_off
```

```python
# ads count
for k,v in user_ads_off.items():
    print("{0}: {1}".format(k, len(v)))
```

```python
# unique ads
for k,v in user_ads_off.items():
    print("{0}: {1}".format(k, len(set(v))))
```

```python
# Top 5 popular ads
from collections import Counter

for k,v in user_ads_off.items():
    if v:
        counter = Counter(v)
        top5 = counter.most_common(5)
        print(k)
        top5_sum = 0
        for i in range(len(top5)):
            print("{0}: {1}".format(top5[i][0], top5[i][1]))
            top5_sum += top5[i][1]
        print("({:.2%})".format(top5_sum/len(v)))
        print('---')
```

```python
# exclusive ads

for worker in tqdm(workers):
    exclusive_ads_count = 0
    print(worker)
    if user_ads_off[worker]:
        for ad in user_ads_off[worker]:
            exclusive = True
            for worker2 in workers:
                if not exclusive:
                    break
                if worker2 == worker:
                    continue
                ads2 = list(set(user_ads_off[worker2]))
                for ad2 in ads2:
                    if ad == ad2:
                        exclusive = False
                        break
            if exclusive:
                print(ad)
                exclusive_ads_count += 1
        print("{0}: {1}".format(worker, exclusive_ads_count))
        print("({:.2%})".format(exclusive_ads_count/len(user_ads_off[worker])))
        print('---')
```
